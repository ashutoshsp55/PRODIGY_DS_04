### Model Prediction 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=100,ngram_range=(1,2))
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
X_train_vect = vectorizer.fit_transform(X_train)
X_test_vect = vectorizer.transform(X_test)



```

```python
from sklearn.svm import LinearSVC
```

```python
clf = LinearSVC()
```

```python
model  = clf.fit(X_train_vect, y_train)
```

```python
y_pred = clf.predict(X_test_vect)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
print(clf.score)
```

### Data Visualisation

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5,4))
plt.pie(sentiment_dist,labels=sentiment_dist,autopct='%1.1f%%',startangle=140)
plt.axis('equal')
plt.title('Sentiment Distribution')
plt.show()        
```

```python
plt.figure(figsize=(5,4))
brand_sentiment = data.groupby(['brand', 'sentiment']).size().unstack(fill_value=0).head(10)
brand_sentiment.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Brand-specific Sentiment Analysis')
plt.xlabel('Brand')
plt.ylabel('Number of Tweets')
plt.legend(title='Sentiment')
plt.xticks(rotation=45)
plt.show()
```

```python
from nltk.tokenize import word_tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from collections import Counter
from wordcloud import WordCloud
import nltk

nltk.download('punkt')
nltk.download('vader_lexicon')

data['Tokens'] = data['text'].apply(word_tokenize)

words = [word for tokens in data['Tokens'] for word in tokens]
word_freq = Counter(words)
top_words = word_freq.most_common(20) 

sid = SentimentIntensityAnalyzer()
data['sentiment'] = data['text'].apply(lambda x: sid.polarity_scores(x)['compound'])

wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(dict(top_words))
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Top 20 Most Frequent Words')
plt.axis('off')
plt.show()

plt.figure(figsize=(8, 6))
plt.hist(data['sentiment'], bins=30, color='skyblue', edgecolor='black')
plt.title('Sentiment Distribution')
plt.xlabel('Compound Sentiment Score')
plt.ylabel('Frequency')
plt.show()
```

```python
sentiment_by_category = data.groupby('brand')['sentiment'].mean().head(10)
plt.figure(figsize=(10, 6))
sentiment_by_category.plot(kind='bar', color='skyblue')
plt.title('Average Sentiment by Category')
plt.xlabel('Category')
plt.ylabel('Average Sentiment Score')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
correlation_matrix = data.corr()
plt.figure(figsize=(6, 4))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.title('Correlation between External Factors and Sentiment')
plt.colorbar()
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.show()
```
